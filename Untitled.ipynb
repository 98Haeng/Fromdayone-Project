```python
import warnings
warnings.simplefilter("ignore")

import konlpy
konlpy.__version__
```

```python
from konlpy.corpus import kolaw
kolaw.fileids()
```

```python
c = kolaw.open('constitution.txt').read()
print(c[:40])
```

```python
from konlpy.tag import Okt

okt=Okt()
print(okt.pos("안녕하세요"))
```

```python
item=pd.read_excel('Item Data.xlsx',index_col='Unnamed: 0')
```

```python
item.head()
```

```python
for text in item['상품 간략설명'].fillna('X'):
    item_words=okt.nouns(text)
    print(item_words)
```

```python
item['상품 간략설명'].fillna('x').apply(lambda x:okt.nouns(x))
```

```python
item['상품설명']=item['상품 간략설명'].fillna('x').apply(lambda x:okt.nouns(x))
```

```python
item.head()
```

```python
item.to_excel('item단어추출본.xlsx',index=False)
```
