```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_excel('아이템기반최종데이터.xlsx')
data.head()
```

```python
data.columns
```

```python
data=data[['상품명','카테고리','패턴재질','모양','계절','종합점수','스타일']]
data.head()
```

```python
df=data.drop_duplicates(['상품명'])
df.shape
```

```python
df.head()
```

```python
df.shape
```

```python
df.index
```

```python
df.index=pd.Series(range(1845),index=df.index)
print(df.index)
```

```python
count_vector = CountVectorizer(ngram_range=(1,1))
```

```python
c_vector1=count_vector.fit_transform(df['카테고리'])
c_vector2=count_vector.fit_transform(df['패턴재질'])
c_vector3=count_vector.fit_transform(df['모양'])
c_vector4=count_vector.fit_transform(df['계절'])
c_vector5=count_vector.fit_transform(df['스타일'])
```

```python
df.dtypes
```

```python
genre_c_sim1=cosine_similarity(c_vector1,c_vector1).argsort()[:,::-1]
genre_c_sim2=cosine_similarity(c_vector2,c_vector2).argsort()[:,::-1]
genre_c_sim3=cosine_similarity(c_vector3,c_vector3).argsort()[:,::-1]
genre_c_sim4=cosine_similarity(c_vector4,c_vector4).argsort()[:,::-1]
genre_c_sim5=cosine_similarity(c_vector5,c_vector5).argsort()[:,::-1]
```

```python
genre_c_sim1
```

```python
genre_c_sim1.shape, genre_c_sim2.shape, genre_c_sim3.shape, genre_c_sim4.shape, genre_c_sim5.shape
```

```python
def get_recommend_item_list1(df,상품이름,top=1845):
    #특정 영화와 비슷한 영화를추천해야 하기에 '특정 영화' 정보를 추출
    target_item_index=df[df['상품명']==상품이름].index.values
    
    #코사인 유사도 중 비슷한 코사인 유사도를 가진 정보 추출
    sim_index1=genre_c_sim1[target_item_index,:].reshape(-1)
    sim_index2=genre_c_sim2[target_item_index,:].reshape(-1)
    sim_index3=genre_c_sim3[target_item_index,:].reshape(-1)
    sim_index4=genre_c_sim4[target_item_index,:].reshape(-1)
    sim_index5=genre_c_sim5[target_item_index,:].reshape(-1)
    
    #본인 제외
    sim_index1=sim_index1[sim_index1 != target_item_index]
    sim_index2=sim_index2[sim_index2 != target_item_index]
    sim_index3=sim_index3[sim_index3 != target_item_index]
    sim_index4=sim_index4[sim_index4 != target_item_index]
    sim_index5=sim_index5[sim_index5 != target_item_index]
    
    #data frame로 만들고 vote_count로 정렬한뒤 return
    #1 카테고리 2패턴재질 3모양 4계절 5스타일
    #우선순위 계절 카테고리 모양 스타일 패턴재질순
    result1=df.iloc[sim_index1].sort_values('종합점수',ascending=False)[:100]
    result2=df.iloc[sim_index2].sort_values('종합점수',ascending=False)[:100]
    result3=df.iloc[sim_index3].sort_values('종합점수',ascending=False)[:100]
    result4=df.iloc[sim_index4].sort_values('종합점수',ascending=False)[:100]
    result5=df.iloc[sim_index5].sort_values('종합점수',ascending=False)[:100]
    
    result4['항목']=5
    result1['항목']=4
    result3['항목']=3
    result5['항목']=2
    result2['항목']=1
    
    results=pd.concat([result1,result2,result3,result4,result5],ignore_index=True)
    results=results.sort_values('종합점수',ascending=False)[:100]
#     results=results.sort_values(['종합점수','계절','카테고리','모양','스타일','패턴재질'],ascending=False)[:1000]
    results=results.drop_duplicates(['상품명'])
    #5개의 result에서 공통된 값들만 합침
    return results
```
